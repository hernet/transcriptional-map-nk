#### Load data

```python
adata = sc.read(os.path.join(data_dir, 'glioblastoma.h5ad'))
```

#### QC

```python
adata.var['mt'] = adata.var_names.str.startswith('MT-')  # annotate mitochondrial genes as 'mt'
adata.var['ribo'] = adata.var_names.str.startswith(("RPS","RPL")) # annotate ribosomal genes as 'ribo'
sc.pp.calculate_qc_metrics(adata, qc_vars=['mt', 'ribo'], percent_top=None, log1p=False, inplace=True)
```

```python
adata = adata[adata.obs.total_counts > 300, :]
adata = adata[adata.obs.pct_counts_mt < 20, :]
adata = adata[adata.obs.total_counts < 40000, :]
```

#### Train model

```python
scvi.model.SCVI.setup_anndata(adata, batch_key="batch", labels_key='cell_type_moreno')
```

```python
vae = scvi.model.SCVI(adata, n_layers=2, n_latent=30, gene_likelihood="nb")
```

```python
vae.train()
```

```python
plt.plot(vae.history["elbo_train"], label="train")
#plt.plot(vae.history["elbo_validation"], label="val")
plt.title("Negative ELBO over training epochs")
#plt.ylim(1100, 1500)
plt.legend()
```

```python
lvae = scvi.model.SCANVI.from_scvi_model(
    vae,
    adata=adata,
    unlabeled_category="Unknown",
)
```

```python
lvae.train(25)
```

```python
adata.obsm["X_scVI"] = vae.get_latent_representation(adata)
```

```python
adata.obsm["X_scANVI"] = lvae.get_latent_representation(adata)
```

```python
sc.pp.neighbors(adata, use_rep="X_scANVI")
```

```python
sc.tl.umap(adata)
```

```python
scanvi_path = 'glioblastoma_tumor_all_cells_scanvi'
```

```python
lvae = scvi.model.SCANVI.load(scanvi_path, adata=adata, use_gpu=True)
```

```python
adata.obs["C_scANVI"] = lvae.predict(adata)
```

```python
sc.pp.neighbors(adata, use_rep="X_scANVI")
sc.tl.umap(adata)
```

```python
cell_type_dict = {'astrocyte' : 'Glial/neuronal',
'dendritic cell' : 'Myeloid',
'endothelial cell' : 'Vascular',
'macrophage' : 'Myeloid',
'malignant cell' : 'Malignant',
'microglial cell' : 'Myeloid',
'monocyte' : 'Myeloid',
'mural cell' : 'Vascular',
'neuron' : 'Neuronal',
'neutrophil' : 'Myeloid',
'oligodendrocyte' : 'Glial/neuronal',
'oligodendrocyte precursor cell' : 'Glial/neuronal',
'radial glial cell' : 'Glial/neuronal'}
```

```python
lymphoid_cell_types = ['B cell', 'mast cell', 'mature T cell','natural killer cell', 'plasma cell']
```

```python
adata.obs['level1_cell_type'] = ['Lymphoid' if (x in lymphoid_cell_types or y in lymphoid_cell_types) else cell_type_dict[y] for x,y in zip(adata.obs['C_scANVI'], adata.obs['cell_type_moreno'])]
```

```python
level2_dict = {'Classical monocytes' : 'Monocytes',
 'astrocyte' : 'Astrocytes', 'endothelial cell' : 'Endothelial cells',
       'malignant cell' : 'Malignant cells', 'microglial cell' : 'Microglial cells', 'monocyte' : 'Monocytes', 'mural cell' : 'Mural cells',
       'neuron' : 'Neurons', 'neutrophil' : 'Neutrophils', 'oligodendrocyte' : 'Oligodendrocytes',
       'oligodendrocyte precursor cell' : 'Oligodendrocytes', 'radial glial cell' : 'Radial glial cells'}
```

```python
adata.obs['level2_cell_type'] = [x if not x in level2_dict else level2_dict[x] for x in adata.obs['level2_cell_type']]
```
