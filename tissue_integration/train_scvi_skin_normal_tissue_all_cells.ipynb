#### Load data

```python
adata = sc.read(os.path.join(data_dir, 'skin_normal_tissue_concatenate_raw.h5ad'))
```

#### QC

```python
adata.var['mt'] = adata.var_names.str.startswith('MT-')  # annotate mitochondrial genes as 'mt'
adata.var['ribo'] = adata.var_names.str.startswith(("RPS","RPL")) # annotate ribosomal genes as 'ribo'
sc.pp.calculate_qc_metrics(adata, qc_vars=['mt', 'ribo'], percent_top=None, log1p=False, inplace=True)
```

```python
adata = adata[adata.obs.total_counts > 300, :]
adata = adata[adata.obs.pct_counts_mt < 20, :]
adata = adata[adata.obs.total_counts < 50000, :]
```

#### Train model

```python
adata = adata.copy()
```

```python
scvi.model.SCVI.setup_anndata(adata, batch_key="batch")
```

```python
vae = scvi.model.SCVI(adata, n_layers=2, n_latent=40, gene_likelihood="nb")
```

```python
vae.train()
```

```python
plt.plot(vae.history["elbo_train"], label="train")
#plt.plot(vae.history["elbo_validation"], label="val")
plt.title("Negative ELBO over training epochs")
#plt.ylim(1100, 1500)
plt.legend()
```

```python
adata.obsm["X_scVI"] = vae.get_latent_representation(adata)
```

```python
sc.pp.neighbors(adata, use_rep="X_scVI")
```

```python
sc.tl.umap(adata)
```

```python
sc.tl.leiden(adata,resolution=0.1) ## 0.3
```

```python
vae.save('skin_normal_tissue_all_cells_scvi', overwrite=True)
```
